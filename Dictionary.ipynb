# Dictionary Characteristics

```python
# Dictionary ordered collection
details ={"Name": "Honey", "Course":"Python", "Batch":57, "age":20}
print(details)
```

```python
# Dictionary values access by using keys
# Uptdating dictionary values
# adding new items to dictionary
details ={"Name": "Honey", "Course":"Python", "Batch":57, "age":20}
print(details['Course'])
# print(details['Location']) #KeyError: 'Location'
# updating dictionary values
details[ 'age'] = 20
print(details['age'])
# Adding new items
details['Location' ] = "Hyderabad"
print(details)
```

```python
details ={'age' :20,'Name': 'Honey, 'Course':'Python', 'Batch':57, 'age':20, 'Location':Hyderabad }
print(details)
```

### Dictionary Methods
-get(key,default_value):It returns the value if key is exists in dictionary,otherwise it returns default value.


```python
## Updating Dictionary
details ={"Name": "Honey", "Course":"Python", "Batch":57, "age":20}
details.update({'batch':'58','course':'java'})
```

### Item removing methods
- popitem(): It returns and removes the last item as tuple
- pop(key): This method removes the specified key itemand returns the specified key value,otherwise KeyError

```python
details = {'age':20, 'name': 'Honey', 'Course': 'java', 'batch':57, 'Location': 'Hyd'}
item = details.popitem()
print(item)
print(details)
val = details.pop('name')
print(val)
print(details)
```

```python
# Membership
# it checks with dictionary keys
details = {'age':20, 'name': 'Honey', 'Course': 'java', 'batch':57, 'Location': 'Hyd'}
print('Honey' in details)
print('name' in details)
```
