```python
# string to int,float,bool,list,tuple,set
# if string value is number
s = '123456'
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# string to int,float,bool,list,tuple,set
# if string value is  float number
s = '123456.1205'
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# list to str,tuple,set
# if list value is string number
s =['246810,13579']
print(str(s))
print(tuple(s))
print(set(s))
```

```python
#tuple to str,list,set
# if tuple value is string number
t=('1345678.12573')
print(str(t))
print(list(t))
print(set(t))
```

```python
# set to str,list,tuple
#it set value is string number
s={2,4,6,8,.10,12}
print(str(s))
print(list(s))
print(tuple(s))
```

```python
# dict to str,list ,set,tuple
# if dict value is str number
d= {'H':'1234.67890.67543'}
print(str(d))
print(list(d))
print(set(d))
print(tuple(d))
```

```python
# list to dictionary conversion
l = [('a',1),('b',2)]
print(dict(l))
```
