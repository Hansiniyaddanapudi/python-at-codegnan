```python
# characteristics
# ordered collection
t = (1,2,3,[10,30,40],23)
print(t)
```

```python
# immutable
# After operation it doesn't change the length of object abd doesn't change thr remaining element position
t = (1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```

```python
# Empty tuple representation
# Single element representation
t1 = ()
t2 = tuple()
print(type(t1),type(t2))
```

```python
# Single element representation
t1 = (10)
t2 = (10,)
l1 = [10]
print(type(t1),type(t2),type(l1))
```

```python
(18,) * 8
```

```python
# tuple operations
t = (10, 20, 30, 40)
# Accessing
print(t[1])       
# Slicing
print(t[1:3])      
# Concatenation
print(t + (40, 50))
# Repetition
print(t * 2)
# Membership
print(30 in t)
# Count
print(t.count(20))
# Index
print(t.index(30))
# Length
print(len(t))

```

```python
#Tuple concatenation
t1 = (18, 23, 84)
t2 = (5, 7)
result = t1 + t2
print(result)
```

```python
#Tuple repetition
colors = ("Red","Black","Blue")
print(colors * 3)
```

```python
# Tuple Methods
# index()
t = (10, 20, 30, 20)
print(t.index(20))
# count()
t = (33, 18, 56, 18, 45)
print(t.count(18))
```
