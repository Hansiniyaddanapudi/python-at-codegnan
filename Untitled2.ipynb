```python
import math
print(math.ceil(4.25)) #It returns the Upperbound integer value
print(math.floor(4.25)) # It returns the lowerbound integer value
print(math.factorial(5)) # It returns the factorial of a number
print(math.gcd(10, 6, 4)) # It returns the Greatest Common Divisor
print(math.lcm(10, 6, 4 )) # It returns the Least Common Multiplier
print(math.pow(10, 3 )) # Pow(a, b)It returns the Float type of value  raise to b
print(math.sqrt(12)) # It returns the squareroot value of a number
print(math.log10(4)) # It Find the log 10 value 
```

```python
# Who will reach top in less steps in dice
count = rolls = 0
print("Welcome to the Who will REach Top In Less Steps Game")
while True:
    input("To Throw a dice, please click enter button")
    face_val = random.randint(1, 6)
    
    if count + face_val <= 100:
        count += face_val
    rolls+=1
    if count == 100:
        print(f"Your Dice value:{face_val}")
        print(f"You reached the Top in {rolls} chances")
        break
        
    print(f"Your Dice value:{face_val}")
    print(f"Your at {count} Position in {rolls} chances")
    print("*" * 50)
```
