### Set Characteristics

```python
# Set unordered collection
# Set return unique elements
s = {1,2,3,2,3,'hello', 'a', 2.02,(10,20,30)}
print(s)
print(s[0])
# Set does not supports indexing
```

```python
s = {1,2,[3,2,3],'hello', 'a', 2.02,(10,20,30)}
print(s)
```

```python
### Set methods
# Element adding method
# add(ele): It adds the element to set if element not in set,otherwise nothing
s= {1,2,3,4,4.3}
print(s)
s.add(100)
s.add("Code")
print(s)
```

```python
# Element removing methods
# Remove(item): It removes the item if it exists in it othewise key error
# Discard(item): It removes the item if it exists in it othewise nothing
```

```python
s ={1,2,3,4,5}
s.remove(5)
print(s)
# s.remove(100)
s.discard(3)
print(s)
s.discard(100)
print(s)
```

```python
# sets Operations
# union(set): It returns the all unique elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
union =s1.union(s2,s3)
print(union)
res = s1|s2|s3
print(res)
```

```python
# Intersection
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
intersection =s1.intersection(s2,s3)
print(intersection)
res = s1&s2&s3
print(res)
```

```python
# Difference
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
difference =s1.difference(s2,s3)
print(difference)
res = s1-s2-s3
print(res)
print((s1|s2)-s3)
```

```python
# symmetric difference(set): It results the non common elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
sy_diff = s1.symmetric_difference(s2)
print(sy_diff)
res = s1^s2
print(res)
```

```python
# issubset()
s1 = {1,3,5,7}
s2 = {4,5,6,7}
issubset= s1.issubset(s2)
print(issubset)
```

```python
# issuperset()
s1 = {1,3,5,7}
s2 = {5,7}
issuperset = s1.issuperset(s2)
print(issuperset)
```

```python
# isdisjointset()
s1 = {1,3,5,7}
s2 = {4,5,6,7}
isdisjointset = s1.isdisjoint(s2)
print(isdisjointset)
```

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5}
print(s1.issubset(s2))
print(s2.issuperset(s2))
print(s3.isdisjoint(s2))
print(dir(set))
```

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
s1.intersection_update(s2)
print(s1)
```
