Conditional statements

```python
#check wheather number is divisible by 3 or not
num = int(input())
if num % 3 == 0 :
    print(f'{num} is divisible by 3')
else:
     print(f'{num} is not divisible by 3')
```

```python
#check wheather the number is divisible by 3 and 5 or not
num = int(input())
if num % 3 ==0 & num % 5 == 0:
    print("Is divisible by 3 & 5")
else:
    print("Is not divisible by 3 & 5")
    
```

```python
# check wheather the number is even or odd
num = int(input())
if num % 2 == 0:
    print("even")
else:
    print("odd")
```

```python
# check wheather the number is even or not
num = int(input())
if num % 2 == 0:
    print("even")
else:
    print("Not even")
```

```python
# generate total bill if the total bill is total bill is more then 500 than give 10 % discount other wise 0 % discount
bill = int(input())
if (bill > 500):
    discount = bill * 0.1
    print(f"bill after 10 % discount:{bill - discount}")
else:
    print(f"bill after 0 % discount:{bill}")
```

```python
# check wheather the product is available or not (read product name and stock)
product = input()
stock = int(input())
if stock > 0 :
    print(f"currently {product} : {stock} items available")
else:
    print(f'currently {product} out of stock')
```

```python
# In a shop contains some items along with prices like{'pen':5
products = {'pen':5,'book':20,'pencil':3}
item = input()
quantity = int(input())
if item in products:
    total_amount = products.get(item) * quantity
    print(f'total amount of {quantity} {item} is: {total_amount}')
else:
    print(f"{item} not available")
```

```python
# check weather the number is +ve ,-ve or zero
num = int(input())
if num > 0:
    print(f'{num} is +ve')
elif num < 0:
    print(f'{num} is -ve')
else:
    print(f'{num} is zero')
    
```

```python
# divisible by 3 - Fizz
# divisible by 5 - Buzz
# both = FizzBuzz
num = int(input())
if num % 3 == 0 and num % 5 == 0:
    print('FizzBuzz')
elif num % 5 == 0:
    print('Buzz')
elif num % 3 == 0:
    print('Fizz')
else:
    print('Invalid')
```
