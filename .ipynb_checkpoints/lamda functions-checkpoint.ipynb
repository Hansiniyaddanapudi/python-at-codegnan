```python
def fun(x):
    if x==0:
        return 0
    fun(x-1)
    print(x)
fun(3)
```

```python
# print 1 to n using recursion
def fun(x):
    if x==0:
        return 0
    print(x)
    fun(x-1)
fun(3)
```

```python
# Find the factorial of numbers
def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x*factorial(x-1)
print(factorial(7))
```

```python
# Find the sum of digits in a number
# n = 456 o/p = 50
def sumofDigits(n):
    if n == 0:
        return 0
    return n % 10 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# Count digits in a number
# n = 456 o/p = 50
def sumofDigits(n):
    if n == 0:
        return 0
    return 1 + sumofDigits(n//10)
print(sumofDigits(456))
print(sumofDigits(15346))
```

```python
# find the squere of a number using lambda
s = lambda x:x**2
print(s(10))
```

```python
# find the squeres of list of integers using map and lambda
arr = [1,2,3,4,5]
print(list(map(lambda x: x**2,arr)))
```

```python
arr = [1,2,3,4,5]
print(list(filter(lambda x: x%2==0 ,arr)))
```

```python
# Check wheather the number is even or odd
even_or_odd = lambda x:"even" if x%2==0 else "odd"
print(even_or_odd(18))
```

```python
arr = [1,2,3,4,5]
# 1 - odd
# 2 - even
arr = [1,2,3,4,5]
print(list(map(lambda x:f"{x}-even"if x%2==0 else f"{x} - odd",arr)))
```

```python
# Find the frequency of each character
string = input()
count = {}
for char in string:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1
print(count)
```

```python
# Find the frequency of each character
string = input()
count = {}
for char in string:
        count[char] = count.get(char, 0) + 1
print(count)
```
