```python
# implement even or odd using functions
def even_or_odd (x:int)->str:
    if(x%2==0):
        return "even"
    else:
        return "odd"
    
# function calling block
n = int(input())
print(even_or_odd(n))

```

```python
# prime number 
def prime_number (n:int)->str:
    if n < 2:
        return"not prime"
    for i in range(2, n//2 + 1):   
        if n % i == 0:
            return" not prime"
    return"prime"
# function calling block
n = int(input())
print(prime_number(n))

        
    
```
