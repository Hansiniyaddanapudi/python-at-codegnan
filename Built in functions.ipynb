```python
# find the difference between two strings
s1 = "CODE"
s2 = "GNAN"
n = len(s1)
s = ""
for i in range(n):
    diff = abs(ord(s1[i]) - ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64+diff)
print(s)
        
```

```python
# find bigger and smaller characters from two string
# compare each character to both strings and if s1[i] > s2[i] add s1[i] to bigger and s2[i] to 
# if both same leave it
s1, s2 = map(str, input().upper())
n = len(s1)
big = small = ""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
        big += s1[i]:
        small += s1[i]:
    else:
        big += s2[i]:
        small += s1[i]:
print(big)
print(small)
    
    
```

```python
# map(fun, iterable): It applies the functionality on each element in iterable
# filter(fun, itearble): It filters the iterables, it returns the values which is not none 
# find the squere of list of integers
def squere(x):
    return x**2
lst = [1,2,3,4,5]
print(list(map(squere,lst)))
```

```python
# filter the even numbers from list of integers
def even(x):
    if x%2 == 0:
        return 1
arr = [1,2,3,4,5,6,7]
print(list(filter(even, arr)))
print(list(map(even, arr)))
```
